### Categories with respect to team

### Age $ Salary distribution of each team

```python

grouped_by_team = df.groupby('Team')
team_stats = grouped_by_team.describe()

for team, data in grouped_by_team:
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(data=data, x='Salary', bins=20, kde=True)
    plt.title(f'Salary Distribution - {team}')
    plt.xlabel('Salary')
    plt.ylabel('Count')
    plt.subplot(1, 2, 2)
    sns.histplot(data=data, x='Age', bins=20, kde=True)
    plt.title(f'Age Distribution - {team}')
    plt.xlabel('Age')
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

### Average Salary of each team

```python
grouped_by_team = df.groupby('Team')
average_salary_by_team = grouped_by_team['Salary'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_by_team.index, y=average_salary_by_team.values)
plt.title('Average Salary by Team')
plt.xlabel('Team')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 1) How many are there in each Team and the percentage splitting with respect to the total employees.

### 5)Find if there is any correlation between age and salary , represent it visually.
